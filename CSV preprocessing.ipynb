```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('sample.csv')

# Now you can work with the data in 'df'
print(df)
```

```python
df
```

```python
df.columns
```

```python
none_counts = df.isna().sum()

# Print the counts
print(none_counts)
```

```python
# Calculate the threshold for removing columns (e.g., 50% of rows)
threshold = len(df) * 0.5

# Identify columns that exceed the threshold
columns_to_drop = none_counts[none_counts > threshold].index

# Drop these columns from the DataFrame
df_cleaned = df.drop(columns=columns_to_drop)

df_cleaned
```

```python
df = df_cleaned
```
