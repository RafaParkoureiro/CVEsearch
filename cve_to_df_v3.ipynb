```python
import nvdlib
import pandas as pd
from datetime import datetime, timedelta

#api_key = '1aac5d44-2d62-415e-92a8-ff2a6e251fc0'
api_key = '56d9404c-f27a-453b-9065-d340bed0562f'

# Define the time range (y, m, day)
end_date = datetime.today()
start_date = end_date - timedelta(days=120)

# We can only work with the cve's from 120 days ago till today (nvdlib limitations)
```

```python
# Search for CVEs within the specified time range
cve_details = nvdlib.searchCVE(pubStartDate=start_date, pubEndDate=end_date, key=api_key)
print(len(cve_details))
```

```python
alldata = []
for cve in cve_details:
    data = {
        'id': cve.id,
        'description': cve.descriptions[0].value if cve.descriptions else None,
        'sourceIdentifier': cve.sourceIdentifier,
        'published': cve.published,
        'lastModified': cve.lastModified,
        'vulnStatus': cve.vulnStatus,
        'exploitAdd': cve.exploitAdd if hasattr(cve, 'exploitAdd') else None,
        'actionDue': cve.actionDue if hasattr(cve, 'actionDue') else None,
        'requiredAction': cve.requiredAction if hasattr(cve, 'requiredAction') else None,
        'references': cve.references if hasattr(cve, 'references') else None,
        'cwe': cve.cwe if hasattr(cve, 'cwe') else None,
        'configurations': cve.configurations if hasattr(cve, 'configurations') else None,
        'url': cve.url if hasattr(cve, 'url') else None,
        'cpe': cve.cpe if hasattr(cve, 'cpe') else None,
        'v31score': cve.v31score if hasattr(cve, 'v31score') else None,
        'v30score': cve.v30score if hasattr(cve, 'v30score') else None,
        'v2score': cve.v2score if hasattr(cve, 'v2score') else None,
        'v31vector': cve.v31vector if hasattr(cve, 'v31vector') else None,
        'v30vector': cve.v30vector if hasattr(cve, 'v30vector') else None,
        'v2vector': cve.v2vector if hasattr(cve, 'v2vector') else None,
        'v31severity': cve.v31severity if hasattr(cve, 'v31severity') else None,
        'v30severity': cve.v30severity if hasattr(cve, 'v30severity') else None,
        'v2severity': cve.v2severity if hasattr(cve, 'v2severity') else None,
        'v31exploitability': cve.v31exploitability if hasattr(cve, 'v31exploitability') else None,
        'v30exploitability': cve.v30exploitability if hasattr(cve, 'v30exploitability') else None,
        'v2exploitability': cve.v2exploitability if hasattr(cve, 'v2exploitability') else None,
        'v31impactScore': cve.v31impactScore if hasattr(cve, 'v31impactScore') else None,
        'v30impactScore': cve.v30impactScore if hasattr(cve, 'v30impactScore') else None,
        'v2impactScore': cve.v2impactScore if hasattr(cve, 'v2impactScore') else None,
        'score': cve.score if hasattr(cve, 'score') else None,
        'v31attackVector': cve.v31attackVector if hasattr(cve, 'v31attackVector') else None,
        'v2accessVector': cve.v2accessVector if hasattr(cve, 'v2accessVector') else None,
        'v31attackComplexity': cve.v31attackComplexity if hasattr(cve, 'v31attackComplexity') else None,
        'v2accessComplexity': cve.v2accessComplexity if hasattr(cve, 'v2accessComplexity') else None,
        'v31privilegesRequired': cve.v31privilegesRequired if hasattr(cve, 'v31privilegesRequired') else None,
        'v31userInteraction': cve.v31userInteraction if hasattr(cve, 'v31userInteraction') else None,
        'v31scope': cve.v31scope if hasattr(cve, 'v31scope') else None,
        'v31confidentialityImpact': cve.v31confidentialityImpact if hasattr(cve, 'v31confidentialityImpact') else None,
        'v2authentication': cve.v2authentication if hasattr(cve, 'v2authentication') else None,
        'v2confidentialityImpact': cve.v2confidentialityImpact if hasattr(cve, 'v2confidentialityImpact') else None,
        'v31integrityImpact': cve.v31integrityImpact if hasattr(cve, 'v31integrityImpact') else None,
        'v2integrityImpact': cve.v2integrityImpact if hasattr(cve, 'v2integrityImpact') else None,
        'v31availabilityImpact': cve.v31availabilityImpact if hasattr(cve, 'v31availabilityImpact') else None,
        'v2availabilityImpact': cve.v2availabilityImpact if hasattr(cve, 'v2availabilityImpact') else None
    }
    
    if data['score']!=None: alldata.append(data)
    # We do not save the ones without score

df = pd.DataFrame(alldata)
```
